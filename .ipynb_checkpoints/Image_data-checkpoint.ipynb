```python
import numpy as np
from PIL import Image
import cv2
def get_pix_val(image):
    im = Image.open(image, "r")
    im = im.getdata()
    image = np.array(im, dtype=np.uint8)
    return image
```

```python
import os
```

```python
off_folder = "/Users/gaky/Desktop/efir/off"
files = os.listdir(off_folder)
```

```python
files
```

```python
image_data = np.zeros((107, 921600, 3))
import time
```

```python
start = time.time()
for i in range(len(files)):
    if files[i] != '.DS_Store' and files[i][:5] != 'files':
        image_data[i] = get_pix_val("/Users/gaky/Desktop/efir/off/"+files[i])
stop = time.time()
```
